```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data # Input variables
y0 = iris.target # Output labels as integers
```

```python
from keras.utils import to_categorical
y = to_categorical(y0) # One-hot encoding of output labels
```

```python
import numpy as np
np.random.seed(0)
```

```python
from sklearn.utils import shuffle
X, y0, y = shuffle(X, y0, y)
```

```python
from keras import models
from keras import layers
```

```python
model = models.Sequential()
model.add(layers.Dense(10, activation="relu", input_dim=X.shape[1]))
model.add(layers.Dense(3, activation="softmax"))
```

```python
model.compile(optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"])
```

```python
model.fit(X, y, epochs=500, validation_split=0.2)
```

Let's predict probabilities for the first two examples.

```python
model.predict(X[0:2,:])
```

Now let's fit the same model but this time encoding the data as integers

```python
model2 = models.Sequential()
model2.add(layers.Dense(10, activation="relu", input_dim=X.shape[1]))
model2.add(layers.Dense(3, activation="softmax"))
model2.compile(optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"])
model2.fit(X, y0, epochs=500, validation_split=0.2)
```
