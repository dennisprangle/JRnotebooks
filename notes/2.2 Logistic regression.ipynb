```python
import numpy as np
import matplotlib.pyplot as plt
```

Set number of points in simulated dataset

```python
n = 200
```

Generate data

```python
np.random.seed(0)
X = np.vstack(
    (np.random.normal(-1., 1.0, (n/2, 2)), np.random.normal(1., 1.0, (n/2, 2)))
)
y = np.hstack((np.repeat(0,n/2), np.repeat(1,n/2)))
```

Plot data

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.savefig("logreg_data.pdf")
```

Perform logistic regression

```python
from keras import models
from keras import layers
```

```python
model = models.Sequential()
model.add(layers.Dense(1, input_dim=2, activation="sigmoid"))
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy",
    metrics=["accuracy"])
```

```python
model.fit(X, y, epochs=500, validation_split=0.2)
```

Grid plot

```python
grid_range = np.arange(-3., 3., 0.02)
dgrid = len(grid_range)
Xpred = np.array([(a, b) for a in grid_range for b in grid_range])
prob_pred = model.predict(Xpred)
ypred = np.round(prob_pred)
ypred = ypred.reshape((dgrid, dgrid))
ypred = np.flipud(ypred)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.imshow(ypred, extent=[-3,3,-3,3])
plt.savefig("classification_boundary.pdf")
```

```python
prob_pred = model.predict(Xpred)
ppred = prob_pred.reshape((dgrid, dgrid))
ppred = np.flipud(ppred)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.imshow(ppred, extent=[-3,3,-3,3])
plt.colorbar()
plt.savefig("classification_probability.pdf")
```
