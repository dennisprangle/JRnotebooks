```python
import keras
```

```python
from keras.datasets import boston_housing
(x_train, y_train), (x_val, y_val) = boston_housing.load_data()
```

```python
from keras import models
from keras import layers
model = models.Sequential()
model.add(layers.Dense(1, activation="linear", input_dim=13))
```

```python
model.compile(optimizer="adam", loss="mean_squared_error")
```

```python
boston_hist = model.fit(x_train, y_train, epochs=5000,
    validation_data=(x_val, y_val))
```

```python
import matplotlib.pyplot as plt
history_dict = boston_hist.history
train_loss = history_dict["loss"]
val_loss = history_dict["val_loss"]
epochs = range(1, len(train_loss) + 1)
```

```python
plt.plot(epochs[100:], train_loss[100:], "b-", label="train")
plt.plot(epochs[100:], val_loss[100:], "r-", label="validation")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.savefig("boston_loss.pdf")
```

The validation loss is above the training loss at some points, which might suggest mild overfitting. This doesn't seem to cause any problems as the test loss does not begin to rise.

```python
pred_train = model.predict(x_train)
pred_val = model.predict(x_val)
```

```python
pred_train = model.predict(x_train)
pred_val = model.predict(x_val)
plt.plot(y_train, pred_train, "bo", label="train")
plt.plot(y_val, pred_val, "ro", label="validation")
plt.legend()
plt.xlabel("True value")
plt.ylabel("Prediction")
plt.savefig("boston_pred.pdf")
```

The various outliers expain why the validation loss is sometimes higher, sometimes lower.
