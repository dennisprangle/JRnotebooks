# TensorBoard example on Boston Housing data

```python
import keras
```

```python
from keras.datasets import boston_housing
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
ntrain, ninputs = x_train.shape
```

```python
import matplotlib.pyplot as plt
```

```python
from keras import models
from keras import layers
from keras import optimizers
from keras import regularizers
from keras import initializers
```

```python
callbacks = [
    keras.callbacks.TensorBoard(
        log_dir = "my_log_dir"
    )
]
```

```python
model = models.Sequential()
model.add(layers.Dense(5, activation='relu', input_dim=13, kernel_regularizer=regularizers.l1(0.1)))
model.add(layers.Dense(1, activation='linear', kernel_regularizer=regularizers.l1(0.1)))
model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(x_train, y_train, epochs=5000, validation_data=(x_test, y_test),
                     callbacks=callbacks)
```

```python
history_dict = history.history
train_loss = history_dict['loss']
test_loss = history_dict['val_loss']
epochs = range(1, len(train_loss) + 1)
plt.plot(epochs[10:], train_loss[10:], 'b-')
plt.plot(epochs[10:], test_loss[10:], 'r-')
```

```python
pred_train = model.predict(x_train)
pred_test = model.predict(x_test)
plt.plot(y_train, pred_train, "bo")
plt.plot(y_test, pred_test, "ro")
```
