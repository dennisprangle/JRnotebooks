# Classification on concentric circles data

## Create data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 200 # Number of points in dataset
np.random.seed(0)
r1 = np.random.gamma(10, 0.1, n/2)
r2 = np.random.gamma(30, 0.1, n/2)
phi1 = np.random.uniform(-np.pi,np.pi,n/2)
phi2 = np.random.uniform(-np.pi,np.pi,n/2)
r = np.hstack((r1, r2))
phi = np.hstack((phi1, phi2))
group = np.hstack((np.repeat(0,n/2), np.repeat(1,n/2)))
x0 = r*np.cos(phi)
x1 = r*np.sin(phi)
X = np.stack((x0,x1), axis=1)
y = group
```

## Logistic regression

```python
from keras import models
from keras import layers
```

```python
model = models.Sequential()
model.add(layers.Dense(1, input_dim=2, activation="sigmoid"))
```

```python
model.compile(optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"])
```

```python
model.fit(X, y, epochs=500, validation_split=0.2)
```

We can plot the decision boundary as follows:

```python
grid_range = np.arange(-5., 5., 0.02)
dgrid = len(grid_range)
Xpred = np.array([(a, b) for a in grid_range for b in grid_range])
prob_pred = model.predict(Xpred)
ypred = np.round(prob_pred)
ypred = ypred.reshape((dgrid, dgrid))
ypred = np.flipud(ypred)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.imshow(ypred, extent=[-4, 4, -4, 4])
#plt.savefig("concentric_boundary.pdf") # Uncomment to save pdf (used in notes!)
```

We can also plot the probabilities of the two classes:

```python
prob_pred = model.predict(Xpred)
ppred = prob_pred.reshape((dgrid, dgrid))
ppred = np.flipud(ppred)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.imshow(ppred, extent=[-5, 5, -5, 5])
plt.colorbar()
```

## Features

The obvious feature to use here is distance from the centre i.e.
$$r = \sqrt{x^2 + y^2}$$
I'll also use the angle from horizontal which might improve things a little bit more
$$s = \tan^{-1} (y/x)$$

```python
r = np.sqrt(np.sum(X**2, 1))
s = np.arctan2(X[:,1], X[:,0])
X_features = np.stack((r,s), axis=1)
```

Now let's fit a logistic regression.

```python
model_features = models.Sequential()
model_features.add(layers.Dense(1, input_dim=2, activation="sigmoid"))
model_features.compile(optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"])
model_features.fit(X_features, y, epochs=500, validation_split=0.2)
```

Now let's plot the decision boundary.

```python
grid_range = np.arange(-5., 5., 0.02)
dgrid = len(grid_range)
Xpred = np.array([(a, b) for a in grid_range for b in grid_range])
rpred = np.sqrt(np.sum(Xpred**2, 1))
spred = np.arctan2(Xpred[:,1], Xpred[:,0])
Xpred_features = np.stack((rpred,spred), axis=1)
prob_pred = model_features.predict(Xpred_features)
ypred = np.round(prob_pred)
ypred = ypred.reshape((dgrid, dgrid))
ypred = np.flipud(ypred)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.imshow(ypred, extent=[-4, 4, -4, 4])
```

Pretty good! However a few points near the boundary are misclassified.
