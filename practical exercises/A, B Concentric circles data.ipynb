# Creating concentric circles data

The following code creates a random dataset - roughly two concentric circles - which can be used as a classification problem.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Set number of points in simulated dataset.

```python
n = 200
```

```python
np.random.seed(0)
r1 = np.random.gamma(10, 0.1, n//2)
r2 = np.random.gamma(30, 0.1, n//2)
phi1 = np.random.uniform(-np.pi,np.pi,n//2)
phi2 = np.random.uniform(-np.pi,np.pi,n//2)
r = np.hstack((r1, r2))
phi = np.hstack((phi1, phi2))
group = np.hstack((np.repeat(0,n//2), np.repeat(1,n//2)))
x0 = r*np.cos(phi)
x1 = r*np.sin(phi)
X = np.stack((x0,x1), axis=1)
y = group
from sklearn.utils import shuffle
X, y = shuffle(X, y)
```

We now have input variables X and target class labels y.

These can be plotted as follows.

```python
plt.scatter(X[:,0],X[:,1],c=y)
```
